### **Creation of Liner Regression Models:**

### Median House Value vs Median Income:

```python
# Creating the model
medianIncome_model = linear_model.LinearRegression()

# Training the model
medianIncome_model.fit(X=train[["median_income"]], y=train[["median_house_value"]])
```

```python
# Creating the model
medianIncome_model_test = linear_model.LinearRegression()

# Training the model
medianIncome_model_test.fit(X=test[["median_income"]], y=test[["median_house_value"]])
```

### Median House Value vs Population:

```python
# Creating the model
population_model = linear_model.LinearRegression()

# Training the model
population_model.fit(X=train[["population"]], y=train[["median_house_value"]])
```

```python
# Creating the model
population_model_test = linear_model.LinearRegression()

# Training the model
population_model_test.fit(X=test[["population"]], y=test[["median_house_value"]])
```

### Median House Value vs Number of Households:

```python
# Creating the model
household_model = linear_model.LinearRegression()

# Training the model
household_model.fit(X=train[["households"]], y=train[["median_house_value"]])
```

```python
# Creating the model
household_model_test = linear_model.LinearRegression()

# Training the model
household_model_test.fit(X=test[["households"]], y=test[["median_house_value"]])
```

### **Calculating Intercept, Slope, MSE, MAE:**

##### In the below solution, calculation for **median house value vs median income**, **median house value vs population**, **median house value vs number of households**

```python
class Calculation_TrainData:

    def calculate_trainData():
        result = []

        # Median House Value vs Median Income
        variable_medianIncome = "Median Income"
        intercept_medianIncome = medianIncome_model.intercept_
        slope_medianIncome = medianIncome_model.coef_
        mae_medianIncome = mean_absolute_error(train[['median_house_value']], medianIncome_model.predict(train[['median_income']]))
        mse_medianIncome = mean_squared_error(train[['median_house_value']], medianIncome_model.predict(train[['median_income']]))

        # Median House Value vs Population
        variable_population = "Population"
        intercept_population = population_model.intercept_
        slope_population = population_model.coef_
        mae_population = mean_absolute_error(train[['median_house_value']], population_model.predict(train[['population']]))
        mse_population = mean_squared_error(train[['median_house_value']], population_model.predict(train[['population']]))

        # Median House Value vs Number of Households
        variable_households = "No.of Households"
        intercept_households = household_model.intercept_
        slope_households = household_model.coef_
        mae_households = mean_absolute_error(train[['median_house_value']], household_model.predict(train[['households']]))
        mse_households = mean_squared_error(train[['median_house_value']], household_model.predict(train[['households']]))

        # Result
        result.append([variable_medianIncome, intercept_medianIncome, slope_medianIncome, mae_medianIncome, mse_medianIncome])
        result.append([variable_population, intercept_population, slope_population, mae_population, mse_population])
        result.append([variable_households, intercept_households, slope_households, mae_households, mse_households])
        result_df = pd.DataFrame(result, columns=['Variables', 'Intercept', 'Slope', 'Mean Absolute Value', 'Mean Squared Value'])
        return result_df
    
    def calculate_testData():
        result = []

        # Median House Value vs Median Income
        variable_medianIncome = "Median Income"
        intercept_medianIncome = medianIncome_model_test.intercept_
        slope_medianIncome = medianIncome_model_test.coef_
        mae_medianIncome = mean_absolute_error(test[['median_house_value']], medianIncome_model_test.predict(test[['median_income']]))
        mse_medianIncome = mean_squared_error(test[['median_house_value']], medianIncome_model_test.predict(test[['median_income']]))

        # Median House Value vs Population
        variable_population = "Population"
        intercept_population = population_model_test.intercept_
        slope_population = population_model_test.coef_
        mae_population = mean_absolute_error(test[['median_house_value']], population_model_test.predict(test[['population']]))
        mse_population = mean_squared_error(test[['median_house_value']], population_model_test.predict(test[['population']]))

        # Median House Value vs Number of Households
        variable_households = "No.of Households"
        intercept_households = household_model_test.intercept_
        slope_households = household_model_test.coef_
        mae_households = mean_absolute_error(test[['median_house_value']], household_model_test.predict(test[['households']]))
        mse_households = mean_squared_error(test[['median_house_value']], household_model_test.predict(test[['households']]))

        # Result
        result.append([variable_medianIncome, intercept_medianIncome, slope_medianIncome, mae_medianIncome, mse_medianIncome])
        result.append([variable_population, intercept_population, slope_population, mae_population, mse_population])
        result.append([variable_households, intercept_households, slope_households, mae_households, mse_households])
        result_df = pd.DataFrame(result, columns=['Variables', 'Intercept', 'Slope', 'Mean Absolute Value', 'Mean Squared Value'])
        return result_df

calculation = Calculation_TrainData
trainData = calculation.calculate_trainData()
testData = calculation.calculate_testData()
print("The calculation for median house value vs other 3 variables for train data are displayed below\n")
print(trainData.to_string(index=False))
print("\nThe calculation for median house value vs other 3 variables for test data are displayed below\n")
print(testData.to_string(index=False))
```
