### **Creation of Liner Regression Models:**

### Median House Value vs Median Income:

```python
# Creating the model
medianIncome_model = linear_model.LinearRegression()

# Training the model
medianIncome_model.fit(X=train[["median_income"]], y=train[["median_house_value"]])
```

### Median House Value vs Population:

```python
# Creating the model
population_model = linear_model.LinearRegression()

# Training the model
population_model.fit(X=train[["population"]], y=train[["median_house_value"]])
```

### Median House Value vs Number of Households:

```python
# Creating the model
household_model = linear_model.LinearRegression()

# Training the model
household_model.fit(X=train[["households"]], y=train[["median_house_value"]])
```

### **Calculating Intercept, Slope, MSE, MAE:**

##### In the below solution, calculation for **median house value vs median income**, **median house value vs population**, **median house value vs number of households**

```python
class Calculation:

    def calMedianIncome():
        result = []
        intercept = medianIncome_model.intercept_
        slope = medianIncome_model.coef_
        mae = mean_absolute_error(train[['median_house_value']], medianIncome_model.predict(train[['median_income']]))
        mse = mean_squared_error(train[['median_house_value']], medianIncome_model.predict(train[['median_income']]))
        result.append([intercept, slope, mae, mse])
        result_df = pd.DataFrame(result, columns=['Intercept', 'Slope', 'MAE', 'MSE'])
        return result_df

calculation = Calculation
cMedianIncome = calculation.calMedianIncome()
print("The calculation for median house value vs median income is displayed below\n")
```

### Median House Value vs Population

```python
class CalculatePopulation:

    def calPopulation():
        result = []
        intercept = population_model.intercept_
        slope = population_model.coef_
        mae = mean_absolute_error(train[['median_house_value']], population_model.predict(train[['population']]))
        mse = mean_squared_error(train[['median_house_value']], population_model.predict(train[['population']]))
        result.append([intercept, slope, mae, mse])
        result_df = pd.DataFrame(result, columns=['Intercept', 'Slope', 'MAE', 'MSE'])
        print("The calculation for median house value vs population is displayed below\n")
        print(result_df)

cPopulation = CalculatePopulation
cPopulation.calPopulation()
```

### Median House Value vs Number of Households

```python
class CalculateHouseholds:

    def calHouseholds():
        result = []
        intercept = household_model.intercept_
        slope = household_model.coef_
        mae = mean_absolute_error(train[['median_house_value']], household_model.predict(train[['households']]))
        mse = mean_squared_error(train[['median_house_value']], household_model.predict(train[['households']]))
        result.append([intercept, slope, mae, mse])
        result_df = pd.DataFrame(result, columns=['Intercept', 'Slope', 'MAE', 'MSE'])
        print("The calculation for median house value vs number of households is displayed below\n")
        print(result_df)

cHouseholds = CalculateHouseholds
cHouseholds.calHouseholds()
```
